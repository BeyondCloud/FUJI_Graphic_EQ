```python
frame_period = 10
which_seg = 10
showN = 500
_NFFT_ = 8192
def getAmpPha(src):
    x,fs = sf.read(src)
    X = librosa.stft(x,n_fft=_NFFT_,win_length = 8192, hop_length=int(44100/(1000/frame_period)))
    absX = np.abs(X)
    phaX = X/absX
    return absX,phaX
def getCurve(seg_A,seg_B):
    peaks_A, _ = find_peaks(seg_A, distance=15)
    peaks_B, _ = find_peaks(seg_B, distance=15)
    peaks_A =  np.insert(peaks_A, 0, 0, axis=0)
    peaks_B =  np.insert(peaks_B, 0, 0, axis=0)
    
    c1 = scipy.interpolate.spline(peaks_A,seg_A[peaks_A],range(showN))
    c2 = scipy.interpolate.spline(peaks_B,seg_B[peaks_B],range(showN))
    c1 = np.maximum(c1,0.5)
    curve = c2/c1

    
    return curve
absi,phai = getAmpPha('my_i.wav')
absa,_ = getAmpPha('a_01.wav')
absGT,_ = getAmpPha('my_a.wav')


###################################
seg_A = cp.copy(absi[:showN,which_seg])
seg_GT = cp.copy(absGT[:showN,which_seg])
seg_B = cp.copy(absa[:showN,which_seg])

curve_ = getCurve(seg_A,seg_B)
# plt.plot(seg_B)
seg_A *= curve_
plt.title('/i/')
plt.plot(seg_A)
plt.show()

for i in range(2):
    curve_tmp = getCurve(seg_A,seg_B)

    curve_ *= curve_tmp
    seg_A *= curve_tmp
    plt.title(i)
    plt.plot(seg_A)
    plt.show()

    

curve = np.ones((int(_NFFT_/2)+1,))

curve[:showN] = curve_
# curve[:10]  = 1
out = modify_specEnv(absi,curve)
xx = librosa.istft(out*phai,hop_length=int(44100/(1000/frame_period)))
xx = xx/max(xx)*0.8
sf.write('my_i_to_a01.wav', xx, fs)

plt.title('i_01')
plt.plot(seg_B)
plt.show()
plt.title('GT')
plt.plot(seg_GT)
plt.show()

ipd.Audio(xx, rate=fs)

```

```python
## frame_period = 10
which_seg = 10
showN = 500
_NFFT_ = 8192
def getAmpPha(src):
    x,fs = sf.read(src)
    X = librosa.stft(x,n_fft=_NFFT_,win_length = 8192, hop_length=int(44100/(1000/frame_period)))
    absX = np.abs(X)
    phaX = X/absX
    return absX,phaX
def getCurveLog(seg_A,seg_B):
    peaks_A, _ = find_peaks(seg_A, distance=20)
    peaks_B, _ = find_peaks(seg_B, distance=20)
    peaks_A =  np.insert(peaks_A, 0, 0, axis=0)
    peaks_B =  np.insert(peaks_B, 0, 0, axis=0)
    
    c1 = scipy.interpolate.spline(peaks_A,seg_A[peaks_A],range(showN))
    c2 = scipy.interpolate.spline(peaks_B,seg_B[peaks_B],range(showN))
    c1 = np.maximum(c1,0.5)
    curveLog = np.power(10,c2-c1)

    
    return curveLog
absi,phai = getAmpPha('my_i.wav')
absa,_ = getAmpPha('a_01.wav')
absGT,_ = getAmpPha('my_a.wav')


###################################
seg_A = cp.copy(absi[:showN,which_seg])
seg_GT = cp.copy(absGT[:showN,which_seg])
seg_B = cp.copy(absa[:showN,which_seg])

seg_A_log = np.log10(seg_A)
seg_B_log = np.log10(seg_B)

plt.title('/a log/')
plt.plot(seg_A)
plt.show()

curve_ = getCurveLog(seg_A_log,seg_B_log)

seg_A *=curve_

plt.plot(seg_A)
plt.show()

for i in range(2):
    seg_A_log = np.log10(seg_A)
    curve_tmp = getCurveLog(seg_A_log,seg_B_log)
    curve_ *= curve_tmp
    seg_A *=curve_tmp
    plt.title('iter:'+str(i))
    plt.plot(seg_A)
    plt.show()

# curve_ = np.power(10,curve_)

curve = np.ones((int(_NFFT_/2)+1,))

curve[:showN] = curve_
# curve[:10]  = 1
out = modify_specEnv(absi,curve)
xx = librosa.istft(out*phai,hop_length=int(44100/(1000/frame_period)))
xx = xx/max(xx)*0.8
sf.write('my_i_to_a01_log.wav', xx, fs)

plt.title('i_01')
plt.plot(seg_B)
plt.show()
plt.title('GT')
plt.plot(seg_GT)
plt.show()

ipd.Audio(xx, rate=fs)

```
